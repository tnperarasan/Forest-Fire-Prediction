### Using Algeria Forest Fire DataSet 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
dataset
```

```python
dataset.info()
```

### Data Cleaning

```python
## Missing Values

dataset[dataset.isnull().any(axis=1)]
```

```python
## Data set is converted into two sets based on the Region 
## 1. Bejaia Region Dataset (Region=0)
## 2. Sidi-Bel Abbes Region Dataset (Region=1)

dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1

## Region Feature Data Type is converted into Integer

dataset['Region']=dataset['Region'].astype(int)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
df=dataset
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
## Removing the space in the Columns name

df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
## Converting the Day,Month,Year object feature into Integer Feature

df [['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)

df.info() 
```

```python
## Converting the other feature into Float feature Except Classes

features=[feature for feature in df.columns if df[feature].dtype!='int64']
features
```

```python
for i in features:
    if i!='Classes':
        df[i]=df[i].astype(float)
df.info()
```

```python
## let Store the cleaned Dataset into new CSV file

df.to_csv('Algerian_forest_fires_Cleaned_dataset.csv')
```
